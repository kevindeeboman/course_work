```python
import pandas as pd
import psycopg2
import numpy as np
```

```python
conn = psycopg2.connect(host="localhost", database="postgres")
cur = conn.cursor()
```

```python
cur.execute("""
CREATE TABLE testy(
    id SERIAL PRIMARY KEY,
    int_1 INTEGER NOT NULL,
    int_2 INTEGER NOT NULL,
    int_3 INTEGER NOT NULL)
    ;""")
conn.commit()
```

```python
rent = np.random.randint(0,100,(100000,3))
```

```python
%%time
for row in rent:
    cur.execute(f"""
    INSERT INTO testy (int_1, int_2, int_3) VALUES({row[0]},{row[1]},{row[2]})
    ;""")
    #print(f'Inserted row {row[0]},{row[1]},{row[2]}')

    
conn.commit()
```

```python
cur.execute("""
SELECT * FROM testy
    ;""")
```
