```python
import psycopg2 as pg2
import pandas as pd
```

```python
conn = pg2.connect(database='dvdrental', user='kevindeeboman')
cur = conn.cursor()
```

```python
def show_res(cur, head=False):
    res = cur.fetchall()
    col_names = [desc[0] for desc in cur.description]
    df = pd.DataFrame(data= res, columns=col_names)
    if head == True:
        display(df.head(3))
    else:
        display(df)
```

```python
# SHOW ALL # Shows the parameters of run-time parameters #
cur.execute("""
    SHOW ALL
    ;
    """)
show_res(cur)
# From the list we can look at a specific parameter such as set TIMEZONE #
cur.execute("""
    SHOW TIMEZONE
    ;
    """)
show_res(cur)
```
